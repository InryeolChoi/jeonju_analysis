```python
params = {
    'ServiceKey': api_key,
	# 'dataValue' : '',
	# 'parkLocation' : '',
    'startPage': '1',
    'pageSize': '350'
}
```

```python
response = requests.get(url, params = params)
print(response.status_code)
```

```python
print(response.text)
```

```python
xml_data = response.text

# XML 데이터 파싱
root = ET.fromstring(xml_data)

# 각 레코드를 담을 리스트 초기화
data = []

# XML 파일의 각 <list> 요소를 파싱하여 데이터프레임으로 변환
for list_elem in root.findall('.//list'):
    record_data = {}
    for elem in list_elem:
        record_data[elem.tag] = elem.text
    data.append(record_data)

# 데이터프레임 생성
df = pd.DataFrame(data)
```
