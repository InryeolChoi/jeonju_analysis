## park 데이터 셋
필요한 열만 불러와서 park 데이터셋 사용

```python
park = parkdata[['dataTitle', 'parkArea','parkFact4','parkGubun','parkLoadAddr','posx','posy' ]]
```

```python
park
```

```python
park.info()
```

워도 경도 faoat로 변환 후 결측치 제거
249 -> 242

```python
# 'posx'와 'posy' 열에서 빈 문자열을 NaN으로 변경
park['posx'] = pd.to_numeric(park['posx'], errors='coerce')
park['posy'] = pd.to_numeric(park['posy'], errors='coerce')

# NaN 값이 있는 행을 제거 (필요한 경우)
park = park.dropna(subset=['posx', 'posy'])
```

```python
park
```

```python
park['parkFact4'].unique()
```

```python
park.info()
```

운동기구가 있는 공원을 기준으로 결측치 제거 242 -> 135

```python
# 공백 제거 및 결측치 처리
park['parkFact4'] = park['parkFact4'].replace(' ', pd.NA)
# NaN 값이 있는 행을 제거 (필요한 경우)
park = park.dropna(subset=['parkFact4'])
```

운동기구, 풋살장, 배드민턴장 등 parkFact4값 쉼표로 구분해 각각의 셀에 삽입

```python
# 'parkFact4' 컬럼의 값을 쉼표로 분리하여 확장
expanded_data = park['parkFact4'].str.split(',', expand=True)

# 결과 출력
print(expanded_data)
```

```python
park['parkFact4-1'] = expanded_data[0]
park['parkFact4-2'] = expanded_data[1]
park['parkFact4-3'] = expanded_data[2]
park['parkFact4-4'] = expanded_data[3]
park['parkFact4-5'] = expanded_data[4]
```
