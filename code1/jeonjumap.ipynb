## 지도 불러오기
- 운동시설이 있는 공원에만 마킹

```python
# 유효한 위도(y)와 경도(x)를 사용합니다.
latitude = 35.8242238
longitude = 127.1479532
jeonju_map = folium.Map(location=[latitude, longitude], zoom_start=13)
jeonju_map
```

```python
from folium.features import CustomIcon
```

```python
# 데이터프레임을 반복하여 마커 추가
for _, row in park.iterrows():
    icon = folium.Icon(icon='glyphicon-star', color='green')  # Folium 내장 아이콘 사용
    popup = folium.Popup(f"Title: {row['dataTitle']}", max_width=200)
    folium.Marker(location=[row['posy'], row['posx']], icon=icon, popup=popup).add_to(jeonju_map)
```

```python
jeonju_map
```

- 공원 전부 표시

```python
# 'posx'와 'posy' 열에서 빈 문자열을 NaN으로 변경
parkdata['posx'] = pd.to_numeric(parkdata['posx'], errors='coerce')
parkdata['posy'] = pd.to_numeric(parkdata['posy'], errors='coerce')

# NaN 값이 있는 행을 제거 (필요한 경우)
parkdata= parkdata.dropna(subset=['posx', 'posy'])
```

```python
parkdata.info()
```

```python
# 데이터프레임을 반복하여 마커 추가
for _, row in parkdata.iterrows():
    icon = folium.Icon(icon='glyphicon-star', color='red')  # Folium 내장 아이콘 사용
    popup = folium.Popup(f"Title: {row['dataTitle']}", max_width=200)
    folium.Marker(location=[row['posy'], row['posx']], icon=icon, popup=popup).add_to(jeonju_map)
```
