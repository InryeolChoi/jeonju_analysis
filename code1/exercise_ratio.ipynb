```python
import pandas as pd
result = pd.DataFrame()
```

```python
data = pd.read_csv("../data/운동현황 3개년.csv", encoding='utf-8')
data['연도'] = data['연도'].str.replace('년', '').astype(int)
data.head()
```

```python
region = data[data['특성별(1)'] == '지역별']
region
```

```python
result[['연도', '특성별(1)', '특성별(2)']] = region[['연도', '특성별(1)', '특성별(2)']]

region['일주일에 3번'] = pd.to_numeric(region['일주일에 3번'], errors='coerce')
region['일주일에 4번'] = pd.to_numeric(region['일주일에 4번'], errors='coerce')
region['일주일에 5번'] = pd.to_numeric(region['일주일에 5번'], errors='coerce')
region['일주일에 6번'] = pd.to_numeric(region['일주일에 6번'], errors='coerce')
region['매일'] = pd.to_numeric(region['매일'], errors='coerce')

# 변환 후 연산 수행
result['자주'] = (region['일주일에 3번'].fillna(0) +
                  region['일주일에 4번'].fillna(0) +
                  region['일주일에 5번'].fillna(0) +
                  region['일주일에 6번'].fillna(0) +
                  region['매일'].fillna(0))
```

```python
result
```

```python
# '특성별(2)'별 '자주' 값의 '연도'별 평균 계산
average_by_feature = result.groupby('특성별(2)')['자주'].mean().reset_index()

# 결과에 '연도'와 '특성별(1)' 추가
average_by_feature['연도'] = '평균'
average_by_feature['특성별(1)'] = '지역별'

# 컬럼 순서 재정렬
result = average_by_feature[['연도', '특성별(1)', '특성별(2)', '자주']]
result
```

```python
result.to_csv("../data/전주시_지역별_자주_운동_비율.csv")
```
