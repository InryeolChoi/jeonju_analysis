### 지방세 계산하기
* 총 부과금액과 총 수납금액 계산
* csv로 방출

```python
import pandas as pd
```

```python
jeonju_tax = pd.read_csv("../data/전주시_지방세_징수_현황.csv", encoding='cp949')
jeonju_tax
```

```python
jeonju_tax1 = jeonju_tax[jeonju_tax['시군구명'] == '전주시 완산구']
jeonju_tax2 = jeonju_tax[jeonju_tax['시군구명'] == '전주시 덕진구']
```

```python
print(str(jeonju_tax1.iloc[2, 1]) + "의 지방세")
data1 = jeonju_tax1.iloc[:, 5:7].sum()
data1
```

```python
print(str(jeonju_tax2.iloc[2, 1]) + "의 지방세")
data2 = jeonju_tax2.iloc[:, 5:7].sum()
data2
```

```python
df1 = pd.DataFrame(data1).transpose()
df1.index = ['완산구']
df2 = pd.DataFrame(data2).transpose()
df2.index = ['덕진구']

df = pd.concat([df1, df2])
df
```

```python
df.to_csv("../data/전주시_지방세.csv")
```
