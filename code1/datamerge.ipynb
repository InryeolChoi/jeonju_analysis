```python
import matplotlib.pyplot as plt

# '소계' 행 데이터
labels = soegae.index
sizes = soegae.values

# 원그래프 그리기
plt.figure(figsize=(10, 7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # 원이 동그랗게 보이도록 비율을 같게 조정
plt.title('sogae persent')
plt.show()
```
