## 체육관 시세

```python
jym_data = pd.read_excel("../data/전주시_체육시설_현황.xlsx")
jym_data.head()
```

```python
jym_data.tail()
```

```python
estate_price = pd.read_csv("../data/전주시_개별공시지가.csv", encoding='cp949', low_memory=False)
estate_price.head()
```

```python
# 함수 정의
def split_address(address):
    pattern = re.compile(r'(\d+)-(\d+)')
    match = pattern.search(address)
    if match:
        first_number = match.group(1)
        second_number = match.group(2)
        remaining_address = pattern.sub('', address).strip()
    else:
        # 패턴이 없을 경우
        parts = address.split()
        if parts and parts[-1].isdigit():
            first_number = parts[-1]
            second_number = ''
            remaining_address = ' '.join(parts[:-1])
        else:
            first_number = ''
            second_number = ''
            remaining_address = address
    return pd.Series([remaining_address, first_number, second_number])
```

```python
# '구주소' 변수를 분리하여 새로운 변수 추가
jym_data[['주소', '본번', '부번']] = jym_data['구주소'].apply(lambda x: pd.Series(split_address(x)))

# 결과 출력
jym_data.head()
```

```python
# 함수 정의: '주소'를 분리하여 '시군구'와 '읍면동'을 반환
def split_address(address):
    parts = address.split()
    시군구 = None
    읍면동 = None
    
    for part in parts:
        if part.endswith('구'):
            시군구 = part
        elif part.endswith('동') or part.endswith('가'):
            읍면동 = part
    
    return 시군구, 읍면동

# '주소' 변수를 분리하여 새로운 변수 추가
jym_data[['시군구', '읍면동']] = jym_data['주소'].apply(lambda x: pd.Series(split_address(x)))
```

```python
# 결과 출력
jym_data.head()
```

```python
# '부번' 열을 동일한 타입(str)으로 변환
jym_data['부번'] = jym_data['부번'].astype(str)
estate_price['부번'] = estate_price['부번'].astype(str)

# jym_data와 estate_price 병합
merged_data = pd.merge(jym_data, estate_price, on = ['시군구', '읍면동', '본번', '부번'], how='left')

if '공시지가' in merged_data.columns:
    # '공시지가' 열을 jym_data에 추가
    jym_data['공시지가'] = merged_data['공시지가']
    # 결과 출력
```

```python
jym_data
```

```python
# '공시지가'가 NaN인 행들을 필터링
nan_rows = jym_data[jym_data['공시지가'].isna()]

# 결측값 대체 함수 정의
def fill_missing_price(row, estate_price):
    bonbun = row['본번']
    # 해당 본번을 가진 행들의 공시지가 평균 계산
    mean_price = estate_price[estate_price['본번'] == bonbun]['공시지가'].mean()
    return mean_price

# '공시지가' 결측값을 본번의 평균 공시지가로 대체
jym_data.loc[jym_data['공시지가'].isna(), '공시지가'] = jym_data[jym_data['공시지가'].isna()].apply(
    fill_missing_price, estate_price=estate_price, axis=1)
```
